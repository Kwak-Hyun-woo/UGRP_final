```python
import torch
import torchvision.transforms as transforms

from PIL import Image

from dataset import MPII
from encoder import AutoEncoder

img = MPII()[5]
encoder = AutoEncoder()
encoder.load_state_dict(torch.load("checkpoint/checkpoint_2021_10_09_04:42:23/checkpoint_2021_10_09_04:42:23_epoch_70.pth")['model_state_dict'])
encoder.eval()
```

```python
invTrans = transforms.Compose([ transforms.Normalize(mean = [ 0., 0., 0. ],
                                                     std = [ 1/0.229, 1/0.224, 1/0.225 ]),
                                transforms.Normalize(mean = [ -0.485, -0.456, -0.406 ],
                                                     std = [ 1., 1., 1. ]),
                               ])

to_PIL = transforms.ToPILImage()
```

```python
img_ori = invTrans(img).squeeze()
img_ori_PIL = to_PIL(img_ori)
img_ori_PIL
```

```python
img_recon = encoder(img.unsqueeze(0))
img_recon = invTrans(img_recon).squeeze()
img_recon = to_PIL(img_recon)
img_recon
```

```python
embedding = encoder.encoder(img.unsqueeze(0))
img_recon = encoder.decoder(embedding)
img_recon = invTrans(img_recon).squeeze()
img_recon = to_PIL(img_recon)
img_recon
```
